```python
import pandas as pd
import re

# =========================
# STEP 1: Load Dataset
# =========================

file_path = "/content/job_descriptions.csv"

# Read CSV safely
try:
    df = pd.read_csv(
        file_path,
        sep=",",
        encoding="latin1",
        engine="python",
        on_bad_lines="skip"
    )

    print("Dataset Loaded Successfully")

except Exception as e:
    print("Error Loading Dataset:", e)


# =========================
# STEP 2: Select Important Columns
# =========================

career_df = df[[
    "Job Title",
    "Role",
    "Job Description",
    "Salary Range",
    "Experience"
]].copy()


# =========================
# STEP 3: Clean Text Function
# =========================

def clean_text(text):
    text = str(text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip().lower()


# Apply cleaning
career_df["Job Title"] = career_df["Job Title"].apply(clean_text)
career_df["Role"] = career_df["Role"].apply(clean_text)
career_df["Job Description"] = career_df["Job Description"].apply(clean_text)
career_df["Experience"] = career_df["Experience"].apply(clean_text)
career_df["Salary Range"] = career_df["Salary Range"].astype(str)


# =========================
# STEP 4: Create Search Column
# =========================

career_df["search_text"] = (
    career_df["Job Title"] + " " +
    career_df["Role"] + " " +
    career_df["Job Description"]
)


# =========================
# STEP 5: User Input
# =========================

user_interest = input("Enter career interest: ").lower()


# =========================
# STEP 6: Search Matching Careers
# =========================

result = career_df[
    career_df["search_text"].str.contains(user_interest, na=False)
]


# =========================
# STEP 7: Show Results
# =========================

if len(result) == 0:
    print("No matching careers found.")

else:
    print("\nRecommended Careers:\n")

    for idx, (_, row) in enumerate(result.head(10).iterrows(), start=1):

        title = str(row["Job Title"]).title()
        role = str(row["Role"]).title()
        exp = " ".join(str(row["Experience"]).split()[:8])
        salary = str(row["Salary Range"])

        print(f"\nCareer Recommendation #{idx}")
        print(f"Career Path: {title}")
        print(f"Role: {role}")
        print(f"Salary Range: {salary}")
        print(f"Relevant Experience: {exp}")
        print("-" * 50)
```
