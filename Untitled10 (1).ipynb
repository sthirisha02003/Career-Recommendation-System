```python
import pandas as pd
import re
from difflib import get_close_matches

print("\n====================================")
print("🤖 AI Career Advisor Agent Started")
print("====================================")

print("\nI can help you with:")
print("- Career Suggestions")
print("- Skills Recommendation")
print("- Salary Insights")
print("- Career Guidance")
print("Starting AI Career Advisor Agent...")

# =====================================
# STEP 1: LOAD DATASET
# =====================================

file_path = "/content/job_descriptions.csv"

try:
    df = pd.read_csv(
        file_path,
        sep=",",
        encoding="latin1",
        engine="python",
        on_bad_lines="skip"
    )

    print("Dataset Loaded Successfully")

except Exception as e:
    print("Dataset Error:", e)
    exit()


# =====================================
# STEP 2: SELECT REQUIRED COLUMNS
# =====================================

required_columns = [
    "Job Title",
    "Role",
    "Job Description",
    "Salary Range",
    "Experience"
]

career_df = df[required_columns].copy()


# =====================================
# STEP 3: CLEAN TEXT
# =====================================

def clean_text(text):
    text = str(text)
    text = text.lower()

    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    return text.strip()


# Apply cleaning
for col in ["Job Title", "Role", "Job Description"]:
    career_df[col] = career_df[col].apply(clean_text)

career_df["Experience"] = career_df["Experience"].astype(str)
career_df["Salary Range"] = career_df["Salary Range"].astype(str)


# =====================================
# STEP 4: CREATE SEARCH COLUMN
# =====================================

career_df["search_text"] = (
    career_df["Job Title"] + " " +
    career_df["Role"] + " " +
    career_df["Job Description"]
)


# =====================================
# STEP 5: AI AGENT MEMORY
# =====================================

conversation_history = []


# =====================================
# STEP 6: AI DECISION ENGINE
# =====================================

def ai_advice(user_text):

    user_text = user_text.lower()

    if "data" in user_text:
        return "Recommended Skills: Python, SQL, Excel, Power BI, Statistics"

    elif "ai" in user_text or "machine learning" in user_text:
        return "Recommended Skills: Python, ML, Deep Learning, NLP"

    elif "marketing" in user_text:
        return "Recommended Skills: SEO, Social Media, Analytics"

    elif "finance" in user_text:
        return "Recommended Skills: Financial Modeling, Excel, Accounting"

    elif "coding" in user_text:
        return "Recommended Skills: Python, Java, DSA, Git"

    else:
        return "Recommended Skills: Communication, Problem Solving, Domain Knowledge"


# =====================================
# STEP 7: AI AGENT LOOP
# =====================================

print("\n==============================")
print("AI Career Advisor Agent Ready")
print("==============================")

print("\nType 'exit' to stop the agent")

while True:

    user_input = input("\nYou: ")

    if user_input.lower() == "exit":
        print("\nAI Agent: Goodbye!")
        break

    cleaned_input = clean_text(user_input)

    # Save memory
    conversation_history.append(user_input)

    print("\nAI Agent: Thinking...")

    # Match careers
    result = career_df[
        career_df["search_text"].str.contains(cleaned_input, na=False)
    ]

    # Smart fuzzy matching
    if len(result) == 0:

        all_titles = career_df["Job Title"].unique()

        suggestions = get_close_matches(
            cleaned_input,
            all_titles,
            n=5,
            cutoff=0.3
        )

        if len(suggestions) > 0:

            suggestion_pattern = "|".join(suggestions)

            result = career_df[
                career_df["search_text"].str.contains(
                    suggestion_pattern,
                    na=False
                )
            ]

    # =====================================
    # STEP 8: SHOW OUTPUT
    # =====================================

    if len(result) == 0:

        print("\nAI Agent:")
        print("No exact match found.")

        print("\nTry keywords like:")
        print("- data analyst")
        print("- software")
        print("- marketing")
        print("- finance")
        print("- ai engineer")

    else:

        print("\n===== Career Recommendations =====")

        for idx, (_, row) in enumerate(result.head(5).iterrows(), start=1):

            print(f"\nRecommendation #{idx}")
            print("Career Path:", str(row["Job Title"]).title())
            print("Role:", str(row["Role"]).title())
            print("Salary Range:", row["Salary Range"])
            print("Experience:", row["Experience"])

            advice = ai_advice(user_input)

            print("\nAI Advice:")
            print(advice)

            print("-" * 50)

    # =====================================
    # STEP 9: MEMORY DISPLAY
    # =====================================

    print("\nConversation Memory:")
    print(conversation_history[-3:])
```
